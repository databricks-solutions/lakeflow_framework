### Get Pipeline Settings

```python
from constants import DLTPipelineSettingKeys

def get_required_config(key: str, description: str) -> str:
    """Get a required config value from spark.conf or raise an error."""
    value = spark.conf.get(key, None)
    if value is None:
        raise ValueError(f"Pipeline settings error: {description} is not set.")
    return value

framework_source_path = get_required_config(
    DLTPipelineSettingKeys.FRAMEWORK_SOURCE_PATH, "Framework source path"
)
```

### Lakeflow Framework Entry Point

```python
import sys

# Append the framework path to the system path
if framework_source_path not in sys.path:
    sys.path.append(framework_source_path)

# Import and initialize the Spark Declarative Pipeline builder
from dlt_pipeline_builder import DLTPipelineBuilder
DLTPipelineBuilder(spark, dbutils).initialize_pipeline()
```
